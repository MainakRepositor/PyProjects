```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
%matplotlib inline
from psutil import cpu_percent
from matplotlib.animation import FuncAnimation
```

```python
cpu_percent()
```

```python
%matplotlib notebook
y = []
frame_length=200
fig = plt.figure(figsize=(10,5))
def animate(i):
    y.append(cpu_percent())
    if len(y) <= frame_length:
        plt.cla()
        plt.plot(y,'r',label = 'Real-time CPU usage')
    
    elif len(y) <= (frame_length + 200):
        plt.cla()
        plt.plot(y[-frame_length:],'y',label = 'Real-time CPU usage')
        
    elif len(y) <= (frame_length + 400):
        plt.cla()
        plt.plot(y[-frame_length:],'g',label = 'Real-time CPU usage')
        
    else:
        plt.cla()
        plt.plot(y[-frame_length:],'cyan',label = 'Real-time CPU usage')
        
    plt.ylim(0,100)
    plt.ylabel('CPU Usage(%)')
    plt.xlabel('Time(s)')
    plt.legend(loc='upper right')
    plt.tight_layout()

ani = FuncAnimation(plt.gcf(),animate,interval=100)
    
```
